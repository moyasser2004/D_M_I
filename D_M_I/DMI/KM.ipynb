### Kmedoid Clustering

```python
import numpy as np
from sklearn_extra.cluster import KMedoids            # type: ignore

list=[[9,6],[10,4],[4,4],[5,8],[3,8],[2,5],[8,5],[4,6], [8,4], [9,3]]

data = np.array(list)

k=2
kmedoids=KMedoids(n_clusters=k).fit(data)
clusters=kmedoids.cluster_centers_
labels= kmedoids.labels_


print("Labels :", labels, "\n")
print("clusters", clusters, "\n")

for j in range (k):
    for i in range (len(data)):
        if (kmedoids.labels_[i]==j):
            x=data[i]
            print(' cluster', j , ':',x)
```

```python
import numpy as np
from sklearn_extra.cluster import KMedoids                 # type: ignore
import pandas as pd

list = pd.read_csv('https://raw.githubusercontent.com/LearnDataSci/glossary/main/data/protein.csv')
list=pd.DataFrame(list)
list=list.drop("Country", axis=1)
list.head()
```

```python
list.isnull().sum()
list.dropna()
```

```python
list.duplicated().sum()
list.drop_duplicates()
```

```python
list=list.drop_duplicates()
print("**Data_set without duplicate rows\n", list,"\n")
```

```python
data=np.array(list)


k=2
kmedoids=KMedoids(n_clusters=k).fit(data)
clusters=kmedoids.cluster_centers_
labels= kmedoids.labels_
print("Labels :", labels, "\n")
print("clusters", clusters, "\n")
for j in range (k):
    for i in range (len(data)):
        if (kmedoids.labels_[i]==j):
            x=data[i]
            print("cluster", j , ":", x)
```
