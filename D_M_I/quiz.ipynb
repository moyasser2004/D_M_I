```python
import random
```

```python
items = [[3, 4], [4, 5], [2, 3], [5, 6], [1, 2]]
print(items)
```

```python
max_weight = 10
population_size = 100
generations = 100
```

```python
# QA  this generate a random population
def generate_population(size):
    population = []
    for i in range(size):
        genes = [0, 1]
        chromosome = []
        for i in range(len(items)):
            chromosome.append(random.choice(genes))
        population.append(chromosome)
    print("Generated a random population of size", size)
    return population


my_population = generate_population(population_size)
my_population
```

```python
# QB   calculate the fitness of a chromosome
def calculate_fitness(chromosome):
    total_weight = 0
    total_value = 0
    for i in range(len(chromosome)):
        if chromosome[i] == 1:
            total_weight += items[i][0]
            total_value += items[i][1]
    if total_weight > max_weight:
        return 0
    else:
        return total_value
```

```python
#  QC  select two chromosomes for crossover
def select_chromosomes(population):
    fitness_values = []
    for chromosome in population:
        fitness_values.append(calculate_fitness(chromosome))

    fitness_values = [float(i) / sum(fitness_values) for i in fitness_values]
    parent1 = random.choices(population, weights=fitness_values, k=1)[0]
    parent2 = random.choices(population, weights=fitness_values, k=1)[0]
    return parent1, parent2
```

```python
def generate_algorithm(population_size, num_generation):
    population = generate_population(population_size)
    for chromosomes in population:
        print(chromosomes)
    for generations in range(num_generation):
        population = select_chromosomes(population)
        print(generations+1)
        for chromosomes in population:
            print(chromosomes)
    best_s = max(my_population, key=calculate_fitness)
    best_f = calculate_fitness(best_s, )
    return best_s, best_f
```

```python
best_s, best_f = generate_algorithm(population_size, generations)
print("Best Fetness: ", best_f)
print("Best Solution: ", best_s)
```
